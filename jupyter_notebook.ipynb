```python
import pandas as pd
df = pd.read_csv('Glassdoor_Scrapped_jobs.csv')
```

```python
df.head()
```

```python
df1 = df[df['Salary Estimate']!='-1']
df1['Salary Estimate'] = df1['Salary Estimate'].apply(lambda x:x.split('(')[0])
df1['Salary Estimate'] = df1['Salary Estimate'].apply(lambda x : 1 if '/hr' in x.lower() else x)
df2 = df1[df1['Salary Estimate']!=1]
```

```python
df2.head()
```

```python
df2.size()
```

```python
df2.Location.value_counts()
```

```python
df2['Job Description'][2]
```

```python
df2['python_yn'] = df2['Job Description'].apply(lambda x:1 if 'python' in x.lower() else 0)
df2.python_yn.value_counts()
```

```python
df2['spark_yn'] = df2['Job Description'].apply(lambda x:1 if 'spark' in x.lower() else 0)
df2.spark_yn.value_counts()
```

```python
df2['r_studio_yn'] = df2['Job Description'].apply(lambda x:1 if 'aws' in x.lower() else 0)
df2.r_studio_yn.value_counts()
```
